```python
import numpy as np
import pandas as pd
from sklearn. model_selection import train_test_split
from sklearn.metrics import recall_score, precision_score, roc_auc_score, accuracy_score, f1_score

from catboost import CatBoostClassifier
```

### Задание 1

Взять любой набор данных для бинарной классификации (можно скачать один из модельных с https://archive.ics.uci.edu/ml/datasets.php)

```python
columns = ['bkblk','bknwy','bkon8','bkona','bkspr','bkxbq','bkxcr','bkxwp','blxwp','bxqsq','cntxt','dsopp','dwipd',
           'hdchk','katri','mulch','qxmsq','r2ar8','reskd','reskr','rimmx','rkxwp','rxmsq','simpl','skach','skewr',
            'skrxp','spcop','stlmt','thrsk','wkcti','wkna8','wknck','wkovl','wkpos','wtoeg', 'white_can_win']


df = pd.Cov = pd.read_csv('../ML_в_бизнесе/kr_vs_kp_data.csv', names=columns)

df.head(11)
```

```python
df['white_can_win'] = df['white_can_win'].replace({'won': 1, 'nowin': 0})
```

```python
df['white_can_win'].value_counts()
```

### Задача 2

Обучить любой классификатор (какой вам нравится)

```python
X = df.iloc[:,:-1]
y = df.iloc[:, -1]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
cat_model = CatBoostClassifier(random_state=42, cat_features=X.columns)

cat_model.fit(X_train, y_train)
y_predict = cat_model.predict(X_test)
```

```python
def evaluate_results(y_test, y_predict):
    f1 = f1_score(y_test, y_predict)
    rec = recall_score(y_test, y_predict, average='binary')
    prc = precision_score(y_test, y_predict, average='binary')

    return round(f1, 2), round(rec, 2), round(prc, 2)
    


f1, recall, precision = evaluate_results(y_test, y_predict)

tab_met = pd.DataFrame({'model': ['CatBoostClassifier'],
                       'fscore': [f1],
                 'precision': [precision],
                 'recall': [recall]})
```

### Задание 3

Разделить ваш набор данных на два множества: P (positives) и U (unlabeled). Причем брать нужно не все положительные примеры (класс 1), а только лишь часть

```python
mod_data = df.copy()
pos_ind = np.where(mod_data.iloc[:, -1].values == 1)[0]

# shuffle them
np.random.shuffle(pos_ind)
# leave just 25% of the positives marked
perc = 0.25
pos_sample_len = int(np.ceil(perc * len(pos_ind)))

print(f'Using {pos_sample_len}/{len(pos_ind)} as positives and unlabeling the rest')
pos_sample = pos_ind[:pos_sample_len]
```

```python
mod_data['class_test'] = -1
mod_data.loc[pos_sample, 'class_test'] = 1
print('target variable:\n', mod_data.iloc[:,-1].value_counts())
```

```python
mod_data.head(10)
```

### Задание 4

Применить random negative sampling для построения классификатора в новых условиях

```python
mod_data = mod_data.sample(frac=1)

data_N = mod_data[mod_data['class_test'] == -1]
data_P = mod_data[mod_data['class_test'] == 1]

neg_sample = data_N[:data_P.shape[0]]
sample_test = data_N[data_P.shape[0]:]
pos_sample = data_P.copy()

print(neg_sample.shape, pos_sample.shape)
sample_train = pd.concat([neg_sample, pos_sample]).sample(frac=1)
```

```python
cat_model = CatBoostClassifier(random_state=42, cat_features=X.columns)
sample_train.loc[sample_train['class_test'] == -1, 'class_test'] = 0

cat_model.fit(sample_train.drop(columns=['class_test', 'white_can_win']),
             sample_train['class_test'])

y_predict = cat_model.predict(X_test)
```

### Задание 5

Сравнить качество с решением из пункта 3 (построить отчет - таблицу метрик)

```python
f1, recall, precision = evaluate_results(y_test, y_predict)

tab_met = tab_met.append({'model': 'CatBoostClassifier + PU learning (25%)',
                       'fscore': f1,
                 'precision': precision,
                 'recall': recall}, ignore_index=True)

tab_met
```

### Задание 6

Поэкспериментировать с долей P на шаге 5 (как будет меняться качество модели при уменьшении/увеличении размера P)

```python
def change_p(p):
    perc = p/100
    pos_sample_len = int(np.ceil(perc * len(pos_ind)))
    pos_sample = pos_ind[:pos_sample_len]

    mod_data['class_test'] = -1
    mod_data.loc[pos_sample, 'class_test'] = 1
    od_data = mod_data.sample(frac=1)

    data_N = mod_data[mod_data['class_test'] == -1]
    data_P = mod_data[mod_data['class_test'] == 1]

    neg_sample = data_N[:data_P.shape[0]]
    sample_test = data_N[data_P.shape[0]:]
    pos_sample = data_P.copy()
    sample_train = pd.concat([neg_sample, pos_sample]).sample(frac=1)
    
    cat_model = CatBoostClassifier(random_state=42, cat_features=X.columns)
    sample_train.loc[sample_train['class_test'] == -1, 'class_test'] = 0

    cat_model.fit(sample_train.drop(columns=['class_test', 'white_can_win']), sample_train['class_test'])

    y_predict = cat_model.predict(X_test) 
    
    return y_predict
    

list_p = [10, 50, 70]

for i in list_p:
    y_pred = change_p(i)
    f1, recall, precision = evaluate_results(y_test, y_pred)
    tab_met = tab_met.append({'model': 'CatBoostClassifier + PU learning (' + str(i) + '%)', 
                              'fscore': f1, 
                              'precision': precision, 
                              'recall': recall}, ignore_index=True)
```
